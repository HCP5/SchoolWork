```python
realOutput=[(2.2,3.1,2.3), (9.7,9.1,2.1), (3.7,4.3,3.9), (5.1,2.7,0), (2.9,5.1,3.2)]
computedOutput=[(2.6,3.1,4.3), (9.1,9.8,3.0), (3.1,5.0,3), (5.1,2.2,0.7), (2.4,5.2,4.2)]

import matplotlib.pyplot as plt
x_cord=[x[0] for x in realOutput]
y_cord=[x[1] for x in realOutput]
z_cord=[x[2] for x in realOutput]
plt.axes(projection='3d')
plt.plot(x_cord,y_cord,z_cord)
plt.show()
```

```python
from math import sqrt
n=(len(realOutput[0]))
Err1=0
for i in range(n):
    Err1+=sum(abs(r[i] - c[i]) for r, c in zip(realOutput, computedOutput)) / len(realOutput)
print("MAE: "+str(Err1/n))
Err2=0
for i in range(n):
    Err2+=sqrt(sum((r[i] - c[i]) ** 2 for r, c in zip(realOutput, computedOutput)) / len(realOutput))
print("RMSE: "+str(Err2/n))
```

```python
from sklearn.metrics import mean_squared_error
import math
toatal=0
for i in range(3):
    forIte=[x[i] for x in realOutput]
    forIteCom=[x[i] for x in computedOutput]
    MSE= mean_squared_error(forIte, forIteCom)
    toatal+=math.sqrt(MSE)
# print(toatal/3)
```
