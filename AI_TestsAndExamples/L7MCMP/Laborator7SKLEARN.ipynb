Reading data
---

```python
import csv
f=open("data.csv","r")
data=csv.DictReader(f)
happScore=[]
capital=[]
freedom=[]
for dataset in data:
    happScore.append(float(dataset['Happiness.Score']))
    capital.append(float(dataset['Economy..GDP.per.Capita.']))
    freedom.append(float(dataset['Freedom']))
print(happScore)
print(capital)
print(freedom)
```

Splitting in train and test cells
---

```python
happScoreTrain=happScore[:125]
capitalTrain=capital[:125]
freedomTrain=freedom[:125]
happScoreTest=happScore[125:]
capitalTest=capital[125:]
freedomTest=freedom[125:]

print("For train: " + str(len(happScoreTrain)))
print("For test: " + str(len(happScoreTest)))
```

Creating regression model from train data
--

```python
from sklearn import linear_model
import numpy as np

capitalT=[[el] for el in capitalTrain]
freedomT=[[el] for el in freedomTrain]

allT=[[x,x2] for x,x2 in zip(capitalTrain,freedomTrain)]
np.array(allT)

regressor=linear_model.LinearRegression()

model=regressor.fit(allT,happScoreTrain)

w0=regressor.intercept_
print(w0)
w1,w2=regressor.coef_

print(regressor.predict(allT))
print(w0+w1*capitalTrain[0]+w2*freedomTrain[0])
```

Prediction
--

```python
inputTest=[[x, y] for x, y in zip(capitalTest, freedomTest)]
computed=model.predict(inputTest)
print("Computed are: ")
print(computed)
print("Real are: ")
print(happScoreTest)
```

Error using RMSE
--

```python
from sklearn.metrics import mean_squared_error


err=mean_squared_error(happScoreTest,computed)
print(err)
```

Error using MAE
--

```python
err+=sum(abs(happScoreTest[i]-computed[i]) for i in range(len(happScoreTest)))
print(err/len(happScoreTest))
```
