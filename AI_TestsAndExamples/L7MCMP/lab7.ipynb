Reading data
---
> `happScore` : hapiness score list <br>
> `capital` : capital List <br>
> `freedom` : freedomIn List <br>

```python
import csv
f=open("data.csv","r")
data=csv.DictReader(f)
happScore=[]
capital=[]
freedom=[]
for dataset in data:
    happScore.append(float(dataset['Happiness.Score']))
    capital.append(float(dataset['Economy..GDP.per.Capita.']))
    freedom.append(float(dataset['Freedom']))
print(happScore)
print(capital)
print(freedom)
```

I'll split the lists into training data and test data

```python
happScoreTrain=happScore[:125]
capitalTrain=capital[:125]
freedomTrain=freedom[:125]
happScoreTest=happScore[125:]
capitalTest=capital[125:]
freedomTest=freedom[125:]

print("For train: " + str(len(happScoreTrain)))
print("For test: " + str(len(happScoreTest)))
```

Making transposed


```python
transX=[]
for c,f in zip(capitalTrain,freedomTrain):
    transX.append([c,f])

for i in range(len(transX)):
    print(transX[i])
```

I'll make the model

```python
n=len(happScoreTrain)
w1=(n*sum([x * y  for x,y in zip(capitalTrain, happScoreTrain)])-sum(capitalTrain)*sum(happScoreTrain))\
    /(n*sum([x**2 for x in capitalTrain]) - sum(capitalTrain)**2)
w2=(n*sum([x * y for x,y in zip(freedomTrain, happScoreTrain)])-sum(freedomTrain)*sum(happScoreTrain))\
    /(n*sum([x**2 for x in freedomTrain]) - sum(freedomTrain)**2)
w0=(sum(happScoreTrain)-w1*sum(capitalTrain)-w2*sum(freedomTrain))\
    /n
print(w1)
print(w2)
print(w0)

testFunc=(w0+w1*capitalTrain[0]+w2*freedomTrain[0])
testFunc2=(w0+w1*capitalTrain[1]+w2*freedomTrain[1])


print(testFunc,happScoreTrain[0])
```

Model Test
--


```python
computed=[]
for i in range(len(happScoreTest)):
    computed.append((w0+w1*capitalTest[i]+w2*freedomTest[i]))
print(computed)
```

Calculating error
--

```python
from math import sqrt
print("RMSE: ")
errorL2 = sqrt(sum((r - c) ** 2 for r, c in zip(happScoreTest, computed)) / len(happScoreTest))
print(errorL2)
print("MAE: ")
errorL1 = sum(abs(r - c) for r, c in zip(happScoreTest, computed)) / len(happScoreTest)
print(errorL1)
```
